```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

iris = load_iris()#load the dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)#create a data frame on data and columns is the feature names
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75 #create a training set
#df['species'] = pd.Factor(iris.target, iris.target_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)#making categorical codes

df.head() #get the first 5 observations

train, test = df[df['is_train']==True], df[df['is_train']==False] #split the data into training andtest

features = df.columns[:4]
clf = RandomForestClassifier(n_jobs=2)#create random forest
y, _ = pd.factorize(train['species']) #taget make the categories into numbers that are fatorcize
clf.fit(train[features], y) #fit the model

preds = iris.target_names[clf.predict(test[features])] #test it
pd.crosstab(test['species'], preds, rownames=['actual'], colnames=['preds'])#check if its right
```

```python
# Decision Tree Classifier
from sklearn import datasets
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
# load the iris datasets
dataset = datasets.load_iris()
# fit a CART model to the data
model = DecisionTreeClassifier()
model.fit(dataset.data, dataset.target)
print(model)
# make predictions
expected = dataset.target
predicted = model.predict(dataset.data)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
